```python
'''
test week4 2022/09/18 22:36
'''
## 採点コード
# common
import sys
import os
sys.path.append(os.path.dirname(os.environ['USERSUBMISSION']))

import warnings
warnings.filterwarnings('ignore')
 
import datetime

score_date = datetime.datetime.now()
score_date = score_date + datetime.timedelta(hours=9)
print(score_date.year, '年', score_date.month, '月',score_date.day, '日',score_date.hour, '時',score_date.minute, '分',score_date.second, '秒時点での提出内容の採点結果です。')

from submission import homework
# common
import pandas as pd
import numpy as np

# init part(データの読み込みと前処理)
FILE_PATH = "score_support.csv" # Omnicampus上にcsvファイルをアップロードしておく
cur_dir = os.path.dirname(__file__)
file_path = os.path.join(cur_dir, FILE_PATH)
online_retail_data_table = pd.read_csv(file_path)


online_retail_data_table['cancel_flg'] = online_retail_data_table.InvoiceNo.map(lambda x:str(x)[0])
# InvoiceNoの先頭が5であるものとIDがNullでないものが対象
target_online_retail_data_tb = online_retail_data_table[(online_retail_data_table.cancel_flg == '5') 
                                                        & (online_retail_data_table.CustomerID.notnull())]
target_online_retail_data_tb = target_online_retail_data_tb.assign(TotalPrice=target_online_retail_data_tb.Quantity * target_online_retail_data_tb.UnitPrice)


# correct answer
def model_answer7(target_online_retail_data_tb, n):

    target_online_retail_data_tb_cbase = target_online_retail_data_tb.groupby("CustomerID")["TotalPrice"].sum()
    target_online_retail_data_tb_cbase.sort_values(ascending=False)
    qcut_result = pd.qcut(target_online_retail_data_tb_cbase, n,  labels=[str(i) for i in range(n)])

    qcut_result_merge = pd.merge(pd.DataFrame(qcut_result),pd.DataFrame(target_online_retail_data_tb_cbase)
         ,right_index=True,left_index=True)
    qcut_result_merge.columns=["qcut_range","total_price"]

    qcut_result_merge_ratio = qcut_result_merge.groupby("qcut_range")["total_price"].sum()/qcut_result_merge.groupby("qcut_range")["total_price"].sum().sum()

    true_result = qcut_result_merge_ratio.sort_index(ascending=False)
    return true_result

def wrong_answer7(target_online_retail_data_tb, n):

    target_online_retail_data_tb_cbase = target_online_retail_data_tb.groupby("CustomerID")["TotalPrice"].sum()
    target_online_retail_data_tb_cbase.sort_values(ascending=False)
    qcut_result = pd.qcut(target_online_retail_data_tb_cbase, n)

    qcut_result_merge = pd.merge(pd.DataFrame(qcut_result),pd.DataFrame(target_online_retail_data_tb_cbase)
         ,right_index=True,left_index=True)
    qcut_result_merge.columns=["qcut_range","total_price"]

    qcut_result_merge_ratio = qcut_result_merge.groupby("qcut_range")["total_price"].sum()/qcut_result_merge.groupby("qcut_range")["total_price"].sum().sum()

    wrong_answer = qcut_result_merge_ratio.sort_values(ascending=False)
    return wrong_answer

def score_homework():
    homework.__globals__['np'] = np
    homework.__globals__['pd'] = pd
    # homework.__globals__['file_url'] = file_url
    # homework.__globals__['online_retail_data'] = online_retail_data
    # homework.__globals__['online_retail_data_table'] = online_retail_data_table
    
    score = 1.0
    for i in [2,5,10,50,70,102,180,200,220]:
        user_result7 = homework(target_online_retail_data_tb, i) # to protect the original data
        true_result7 = model_answer7(target_online_retail_data_tb, i)
        
        # print(user_result7)
        # print(true_result7)
                
        if ((user_result7.values - true_result7.values)/true_result7.values).sum() >= 0.01:
            if ((user_result7.values - wrong_answer7(target_online_retail_data_tb, i).values)/wrong_answer7(target_online_retail_data_tb, i).values).sum() < 0.01:
                print("ヒント：グループ順ではなく、割合の降順になっていませんか？")
            score = 0.0
            break

    print(score)
    #return(score)

if __name__ == '__main__':
    try:
        score_homework()
    except Exception as e:
        print('【ERROR】')
        print(e)
        print(1000)
```

```python
'''
2024summmer 利用コード

'''

## 採点コード
# common
import sys
import os
sys.path.append(os.path.dirname(os.environ['USERSUBMISSION']))

import warnings
warnings.filterwarnings('ignore')
 
import datetime

score_date = datetime.datetime.now()
score_date = score_date + datetime.timedelta(hours=9)
print(score_date.year, '年', score_date.month, '月',score_date.day, '日',score_date.hour, '時',score_date.minute, '分',score_date.second, '秒時点での提出内容の採点結果です。')

print('この「採点結果の出力」の最終行に1.0が出力されており、採点結果点数に1が表示されている場合は正解です。')

from submission import homework
# common
import pandas as pd
import numpy as np

# init part(データの読み込みと前処理)
FILE_PATH = "/home/app/answers/score_support.csv" # Omnicampus上にcsvファイルをアップロードしておく
#cur_dir = os.path.dirname(__file__)
#file_path = os.path.join(cur_dir, FILE_PATH)
online_retail_data_table = pd.read_csv(FILE_PATH)

# url='https://drive.google.com/file/d/1og0u1wN_V_jDwEaz_8wsaZ4avb-qPoTF/view?usp=share_link'
# url='https://drive.google.com/uc?id=' + url.split('/')[-2]
# online_retail_data_table = pd.read_csv(url)


online_retail_data_table['cancel_flg'] = online_retail_data_table.InvoiceNo.map(lambda x:str(x)[0])
# InvoiceNoの先頭が5であるものとIDがNullでないものが対象
target_online_retail_data_tb = online_retail_data_table[(online_retail_data_table.cancel_flg == '5') 
                                                        & (online_retail_data_table.CustomerID.notnull())]
target_online_retail_data_tb = target_online_retail_data_tb.assign(TotalPrice=target_online_retail_data_tb.Quantity * target_online_retail_data_tb.UnitPrice)


# correct answer
def model_answer7(target_online_retail_data_tb, n):
    # 合計購買金額をベースに降順に並び替え
    df = target_online_retail_data_tb.groupby('CustomerID')['TotalPrice'].sum().sort_values(ascending=False)
    # 合計購買金額の降順に並んだ購買者をn等分する
    groups = pd.qcut(df, n).reset_index().rename(columns={'TotalPrice': 'groups'})
    # merge
    df = pd.merge(df.reset_index(), groups, on="CustomerID")
    # グループごとに購買金額の総和を算出
    df = df.groupby("groups")["TotalPrice"].sum() / df.groupby("groups")["TotalPrice"].sum().sum()
    # 降順に並び替える
    my_result = df.sort_values(ascending=False)
    
    return my_result

def score_homework():
    homework.__globals__['np'] = np
    homework.__globals__['pd'] = pd
    # homework.__globals__['file_url'] = file_url
    # homework.__globals__['online_retail_data'] = online_retail_data
    # homework.__globals__['online_retail_data_table'] = online_retail_data_table
    
    score = 1.0
    for i in [2,5,10,50,70,102,180,200,220]:
        user_result7 = homework(target_online_retail_data_tb, i) # to protect the original data
        true_result7 = model_answer7(target_online_retail_data_tb, i)
        
        # print(user_result7)
        # print(true_result7)
                
        if ((user_result7.values - true_result7.values)/true_result7.values).sum() >= 0.1:
            score = 0.0
            break

    print(score)
    #return(score)

if __name__ == '__main__':
    try:
        score_homework()
    except Exception as e:
        print('【ERROR】')
        print(e)
        print(1000)
```
