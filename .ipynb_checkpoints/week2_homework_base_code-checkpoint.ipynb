```python

'''
test_week2
'''
# common
import sys
import os
sys.path.append(os.path.dirname(os.environ['USERSUBMISSION']))

import warnings
warnings.filterwarnings('ignore')
 
import datetime

score_date = datetime.datetime.now()
score_date = score_date + datetime.timedelta(hours=9)
print(score_date.year, '年', score_date.month, '月',score_date.day, '日',score_date.hour, '時',score_date.minute, '分',score_date.second, '秒時点での提出内容の採点結果です。')
print('この「採点結果の出力」の最終行に1.0が出力されており、採点結果点数に1が表示されている場合は正解です。')

from submission import homework
import numpy as np
from numpy import linalg

matrix_list = []
#matrix_list.append(np.array([[5]]))
matrix_list.append(np.array([2, 1, 1, 5]))
matrix_list.append(np.array([0, 3, 9, 15]))
matrix_list.append(np.array([25, 19, 155, 55]))
matrix_list.append(np.array([0, 15555, 35, 13, 93, 95, 90, 32]))

# correct answer
def model_answer2(a):
    idx1 = (a % 2 != 0)
    idx2 = (a % 5 == 0)
    idx = idx1 & idx2
    return a[idx]

def score_homework():
    homework.__globals__['np'] = np
    homework.__globals__['linalg'] = linalg

    score = 1.0
    for i in matrix_list:
        user_result2 = homework(i) # to protect the original data
        true_result2 = model_answer2(i)
        #print(f"matrix:\n {i}")
        #print(f"true result: {true_result2}")
        #print(f"your result: {user_result2}")
        #print(f"error: {np.abs((user_result2 - true_result2)/true_result2)}")
        if np.abs(((user_result2 - true_result2)/true_result2).sum()) >=  1e-7:
            score = 0.0 
            print(f"誤差: {np.abs((user_result2 - true_result2)*100/true_result2):.2f}%")
            print("10^(-5)%以上の誤差があります。")
            break

    print(score)

if __name__ == '__main__':
    try:
        score_homework()
    except Exception as e:
        print('【ERROR】')
        print(e)
        print(1000)
```
