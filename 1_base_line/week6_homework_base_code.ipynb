```python
'''
test_week6_(5月25日授業の宿題) 2024/11/01 08:59
'''
# common
import sys
import os
sys.path.append(os.path.dirname(os.environ['USERSUBMISSION']))
 
import warnings
warnings.filterwarnings('ignore')
 
import datetime
 
score_date = datetime.datetime.now()
score_date = score_date + datetime.timedelta(hours=9)
print(score_date.year, '年', score_date.month, '月',score_date.day, '日',score_date.hour, '時',score_date.minute, '分',score_date.second, '秒時点での提出内容の採点結果です。')
print('この「採点結果の出力」の最終行に1.0が出力されており、採点結果点数に1が表示されている場合は正解です。')
from submission import homework
# common
import pandas as pd
import numpy as np
 

 # init part(データの読み込みと前処理)
FILE_PATH = "/home/app/answers/score_support.csv" # Omnicampus上にcsvファイルをアップロードしておく
#cur_dir = os.path.dirname(__file__)
#file_path = os.path.join(cur_dir, FILE_PATH)
online_retail_data_table = pd.read_csv(FILE_PATH)



online_retail_data_table['cancel_flg'] = online_retail_data_table.InvoiceNo.map(lambda x:str(x)[0])
online_retail_data_table["StockCode"] = online_retail_data_table["StockCode"].astype(str) 

# InvoiceNoの先頭が5であるものとIDがNullでないものが対象
target_online_retail_data_tb = online_retail_data_table[(online_retail_data_table.cancel_flg == '5') 
                                                        & (online_retail_data_table.CustomerID.notnull())]
 
target_online_retail_data_tb = target_online_retail_data_tb.assign(TotalPrice=target_online_retail_data_tb.Quantity * target_online_retail_data_tb.UnitPrice)
 
# correct answer
def model_answer12(target_online_retail_data_tb):
    target_online_retail_data_tb["StockCode"] = target_online_retail_data_tb["StockCode"].astype(str) # 追加 
    online_retail_data_table_first = target_online_retail_data_tb[target_online_retail_data_tb["StockCode"]=="22383"]
    online_retail_data_table_second = target_online_retail_data_tb[target_online_retail_data_tb["StockCode"]=="20725"]
 
    merge_one_second = pd.merge(online_retail_data_table_first
                           ,online_retail_data_table_second
                           ,on="InvoiceNo"
                           ,how="inner")
 
    a = len(merge_one_second.InvoiceNo.unique()) / len(online_retail_data_table_second.InvoiceNo.unique())
    b = len(online_retail_data_table_first.InvoiceNo.unique()) / len(target_online_retail_data_tb.InvoiceNo.unique())
 
    true_result =  a / b
 
    return true_result
 
def score_homework():
    homework.__globals__['np'] = np
    homework.__globals__['pd'] = pd
    #homework.__globals__['file_url'] = file_url
    #homework.__globals__['online_retail_data'] = online_retail_data
    # homework.__globals__['online_retail_data_table'] = online_retail_data_table
    user_result12 = homework(target_online_retail_data_tb) # to protect the original data
    true_result12 = model_answer12(target_online_retail_data_tb)
    if np.abs(user_result12 - true_result12) < 0.01:
        score = 1.0
    else:
        score = 0.0
    print(score)
    #return(score)
 
if __name__ == '__main__':
    try:
        score_homework()
    except Exception as e:
        print('【ERROR】')
        print(e)
        print(1000)
```
