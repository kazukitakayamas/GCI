```python
'''
test_week5_(5月18日授業の宿題)
'''
# common
import sys
import os
sys.path.append(os.path.dirname(os.environ['USERSUBMISSION']))
 
import warnings
warnings.filterwarnings('ignore')
 
import datetime
 
score_date = datetime.datetime.now()
score_date = score_date + datetime.timedelta(hours=9)
print(score_date.year, '年', score_date.month, '月',score_date.day, '日',score_date.hour, '時',score_date.minute, '分',score_date.second, '秒時点での提出内容の採点結果です。')
print('この「採点結果の出力」の最終行に1.0が出力されており、採点結果点数に1が表示されている場合は正解です。')
from submission import homework
# common
import numpy as np
import pandas as pd
from sklearn import linear_model
 
# URL
path_winequality_data = "https://lecturematerials.s3-ap-northeast-1.amazonaws.com/GCI/2021_Summer/homework/winequality-red.csv"
 
column_list = [['volatile acidity', 'quality'], 
               ['chlorides', 'sulphates'], 
               ['citric acid', 'pH'], 
               ['alcohol', 'density'], 
               ['free sulfur dioxide', 'fixed acidity'], 
               ['citric acid', 'total sulfur dioxide',]]
 
# correct answer
def model_answer3(path_winequality_data, X_column, Y_column):
    wine = pd.read_csv(path_winequality_data, sep=";")
 
    lm_reg = linear_model.LinearRegression()
 
    X = wine.loc[:, [X_column]].values
    Y = wine[Y_column].values
 
    lm_reg.fit(X, Y)
 
    true_result = lm_reg.score(X, Y)
    return true_result
 
def score_homework():
    homework.__globals__['np'] = np
    homework.__globals__['pd'] = pd
    homework.__globals__['linear_model'] = linear_model
 
    score = 1.0
    for A in column_list:
        user_result3 = homework(path_winequality_data, A[0], A[1]) # to protect the original data
        true_result3 = model_answer3(path_winequality_data, A[0], A[1])
 
        if np.abs(user_result3-true_result3) >= 0.01:
            score = 0.0
            break
 
    print(score)
    #return(score)
 
if __name__ == '__main__':
    try:
        score_homework()
    except Exception as e:
        print('【ERROR】')
        print(e)
        print(1000)
```
